```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, RandomForestRegressor, AdaBoostRegressor, BaggingClassifier, BaggingRegressor, StackingClassifier, StackingRegressor
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import accuracy_score,balanced_accuracy_score,confusion_matrix, r2_score, mean_absolute_error, recall_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
# Codificação das variáveis categóricas(strings em números)
le = LabelEncoder()
df_t = df.copy(deep=True)
df_t['shipment'] = le.fit_transform(df_t['shipment'])
print(list(le.classes_))
df_t['gender'] = le.fit_transform(df_t['gender'])
print(list(le.classes_))
df_t['product_importance'] = le.fit_transform(df_t['product_importance'])
print(list(le.classes_))
df_t['warehouse'] = le.fit_transform(df_t['warehouse'])
print(list(le.classes_))
df_t
```

```python
# Variáveis numéricas com escala usando Min_Max(ajustar as escalas entre 0 e 1)
mm =  MinMaxScaler()
df_t[['product_cost','weight','discount']] = mm.fit_transform(df_t[['product_cost','weight','discount']])
df_t
```

```python
# Remover um falso preditor
df_t = df_t.drop(columns = ['customer_rating'])
df_t
```

```python
# Agrupar o dataframe pelas colunas week_of_purchase e on_time para calcular a percentagem de 1's em cada semana
percentages = df_t.groupby(['week_of_purchase', 'on_Time']).size().unstack().fillna(0)
percentages = percentages.apply(lambda x: x / x.sum(), axis=1)

# Criação de um gráfico com as percentagens ao longo do tempo
percentages.plot(kind='line', y=1, label='On Time')
plt.title('Percentage of On-Time Purchases over Time')
plt.xlabel('Week of Purchase')
plt.ylabel('Percentage of On-Time Purchases')
plt.show()
```

```python
# Separação do dataset em treino e teste
X = df_t.drop(columns = ['on_Time'])
Y = df_t['on_Time']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
# Função para fazer avaliação do modelo respetivo
def Classification(model,X_train,y_train,X_test,y_test):
    model.fit(X_train,y_train)
    model_predict = model.predict(X_test)
    print(f"{type(model).__name__} Accuracy Score: {accuracy_score(y_test, model_predict)} Balanced Accuracy Score: {balanced_accuracy_score(y_test, model_predict)}")
    print(f"{type(model).__name__} Recall Score: {recall_score(y_test, model_predict,pos_label = 0)} Specificity: {recall_score(y_test, model_predict,pos_label = 1)}")
    print(confusion_matrix(y_test, model_predict))
```

```python
# Classificações para data não balanceada
models = []

models.append(KNeighborsClassifier(n_neighbors=3)) # Knn Algorithm

models.append(MLPClassifier(random_state = 42, max_iter=1000)) # Neural Network

models.append(DecisionTreeClassifier(random_state = 42)) # Decision Trees

models.append(RandomForestClassifier(random_state = 42)) # Random Forest

models.append(LogisticRegression(max_iter=1000)) 

models.append(AdaBoostClassifier(random_state = 42)) # Ada Boost

models.append(BaggingClassifier(random_state = 42)) # Ada Boost

estimators = [('dt', DecisionTreeClassifier(random_state = 42)),('mlp',MLPClassifier(random_state = 42)),('knn', KNeighborsClassifier())]
models.append(StackingClassifier(estimators=estimators, final_estimator=LogisticRegression()))

for model in models:
    Classification(model,X_train,y_train,X_test,y_test)
```

```python
# Cross Validation da data não balanceada
cv = KFold(n_splits=5, shuffle=True, random_state=42)
for model in models:
    cvs = cross_val_score(model, X, Y, cv=cv, n_jobs=4)
    print(f"{type(model).__name__}: {cvs} mean:{np.mean(cvs)}")
```

```python
# Balanceamento da Data
groups = df_t.groupby('on_Time')

# Obter o número mínimo de amostras para cada classe
min_samples = groups.size().min()

# Amostrar o mesmo número de amostras para cada classe
sampled_data = groups.apply(lambda x: x.sample(n=min_samples))

Xb = sampled_data.drop(columns = ['on_Time'])
Yb = sampled_data['on_Time']
Xb_train, Xb_test, yb_train, yb_test = train_test_split(Xb, Yb, test_size = 0.3, random_state = 42) 
```

```python
# Classificações para data balanceada
models = []

models.append(KNeighborsClassifier(n_neighbors=3)) # Knn Algorithm

models.append(MLPClassifier(random_state = 42)) # Neural Network

models.append(DecisionTreeClassifier(random_state = 42)) # Decision Trees

models.append(RandomForestClassifier(random_state = 42)) # Random Forest

models.append(LogisticRegression(max_iter=10000)) 

models.append(AdaBoostClassifier(random_state = 42)) # Ada Boost

models.append(BaggingClassifier(random_state = 42)) # Ada Boost

estimators = [('dt', DecisionTreeClassifier(random_state = 42)),('mlp',MLPClassifier(random_state = 42)),('knn', KNeighborsClassifier())]
models.append(StackingClassifier(estimators=estimators, final_estimator=LogisticRegression()))

for model in models:
    Classification(model,Xb_train,yb_train,Xb_test,yb_test)
```

```python
# Cross Validation da data balanceada
cv = KFold(n_splits=5, shuffle=True, random_state=42)
for model in models:
    cvs = cross_val_score(model, Xb, Yb, cv=cv, n_jobs=4)
    print(f"{type(model).__name__}: {cvs} mean:{np.mean(cvs)}")
```

```python
# Modelo para previsão de vendas
df2 = df_t.groupby(['week_of_purchase']).size().reset_index(name='Purchases')
df2['Purchases'].plot()
plt.show() 
```

```python
# Informação acerca do dataframe
df2
```

```python
# Criar novas colunas com data shifted
df3 = pd.DataFrame({
    'previous_week_2': df2['Purchases'].shift(2),
    'previous_week_1': df2['Purchases'].shift(1),
    'current_week': df2['Purchases'],
    'future_week': df2['Purchases'].shift(-1)
})

# retirar os valores de NaN nas primeiras 2 linhas
df3 = df3.dropna()

# Dar print do dataframe atualizado
print(df3)
```

```python
# Correlação entre as variáveis
sns.heatmap(df3.corr(),annot=True, fmt=".2f")
plt.show()
```

```python
# PairPlot das variáveis
sns.pairplot(df3)
```

```python
# Separação do dataset em treino e teste
X = df3.drop(columns = ['future_week'])
Y = df3['future_week']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
# Função para fazer avaliação do modelo respetivo
def Regression(model):
    model.fit(X_train,y_train)
    model_predict = model.predict(X_test)
    print(f"{type(model).__name__} R^2: {r2_score(y_test, model_predict)}")
    print("MAE:",mean_absolute_error(y_test, model_predict))
```

```python
# Regressões
models = []
models.append(KNeighborsRegressor(n_neighbors=3)) # Knn Algorithm

models.append(MLPRegressor(max_iter=10000,random_state = 42)) # Neural Network

models.append(DecisionTreeRegressor(random_state = 42)) # Decision Trees

models.append(RandomForestRegressor(random_state = 42)) # Random Forest

models.append(LinearRegression())

models.append(AdaBoostRegressor(random_state = 42)) # Ada Boost

models.append(BaggingRegressor(random_state = 42)) # Ada Boost

estimators = [('dt', DecisionTreeRegressor(random_state = 42)),('mlp',MLPRegressor(random_state = 42,max_iter=10000)),('knn', KNeighborsRegressor())]
models.append(StackingRegressor(estimators=estimators)) # Stacking

for model in models:
    Regression(model)
```

```python
# Cross Validation da data
cv = KFold(n_splits=3, shuffle=True, random_state=42)
for model in models:
    cvs = cross_val_score(model, X, Y, cv=cv, n_jobs=4)
    print(f"{type(model).__name__}: {cvs} mean:{np.mean(cvs)}")
```
